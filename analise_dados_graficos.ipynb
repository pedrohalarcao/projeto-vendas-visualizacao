# **Trabalhando com a biblioteca Matplotlib**

##**Conhecendo os dados**

```python
import pandas as pd
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
               'Fev': [120, 90, 170, 60],
               'Mar': [150, 100, 200, 80],
               'Abr': [180, 110, 230, 90],
               'Mai': [220, 190, 350, 200],
               'Jun': [230, 150, 280, 120],
               'Jul': [250, 170, 300, 140],
               'Ago': [260, 180, 310, 150],
               'Set': [240, 160, 290, 130],
               'Out': [220, 140, 270, 110],
               'Nov': [400, 220, 350, 190],
               'Dez': [300, 350, 400, 250],
}
```

###Realizando alguns tratamentos

##**Criando gr√°ficos/figuras com Matplotlib - subplots**

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

df = df_vendas_2022

axs[0, 0].plot(df.loc['A', meses])
axs[0, 0].set_title('Loja A')

axs[0, 1].plot(df.loc['B', meses])
axs[0, 1].set_title('Loja B')

axs[1, 0].plot(df.loc['C', meses])
axs[1, 0].set_title('Loja C')

axs[1, 1].plot(df.loc['D', meses])
axs[1, 1].set_title('Loja D')

plt.show()
```

# Objetivo
Criar um gr√°fico de linhas √∫nico usando seaborn para visualizar os dados de vendas das 4 lojas. Utilizar uma paleta de cores definida por `sns.set_palette()`.

## Prepara√ß√£o dos dados




```python
df_long = df_vendas_2022.stack().reset_index()
df_long.columns = ['Loja', 'M√™s', 'Vendas']
meses_ordenados = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
df_long['M√™s'] = pd.Categorical(df_long['M√™s'], categories=meses_ordenados, ordered=True)
display(df_long.head())
```

##**Personalizando visualiza√ß√µes com Seaborn**

## Definindo paleta de cores




```python
import seaborn as sns

sns.set_palette('tab10')
```

## Criando gr√°fico de linhas




```python
plot = sns.lineplot(data=df_long, x='M√™s', y='Vendas', hue='Loja')
```

## Alterando as caracter√≠sticas do gr√°fico

```python
plt.figure(figsize=(12, 6))
plot = sns.lineplot(data=df_long, x='M√™s', y='Vendas', hue='Loja')
plot.set_title('An√°lise das Vendas Mensais por Loja em 2022', fontsize=16)
plot.set_xlabel('M√™s', fontsize=12)
plot.set_ylabel('Total de Vendas', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##**Gr√°fico animado com Plotly**

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(
    data=[go.Scatter(x=df_long[df_long['Loja']==loja]['M√™s'][:1],
                     y=df_long[df_long['Loja']==loja]['Vendas'][:1],
                     mode='lines',
                     name=loja) for loja in lojas],
    layout=go.Layout(
        title='An√°lise das Vendas Mensais por Loja em 2022',
        xaxis_title='M√™s',
        yaxis_title='Total de Vendas',
        xaxis=dict(range=[-0.5, len(meses_ordenados)-0.5]),
        yaxis=dict(range=[df_long['Vendas'].min(), df_long['Vendas'].max()]),
        hovermode='closest',
        updatemenus=[dict(type="buttons",
                          buttons=[dict(label="Play",
                                        method="animate",
                                        args=[None, {"frame": {"duration": 500, "redraw": False}, "fromcurrent": True, "transition": {"duration": 300, "easing": "quadratic-in-out"}}])])],
    ),
    frames=[]
)
```

```python
frames = []
for k, mes in enumerate(meses_ordenados):
    data = []
    for loja in lojas:
        df_filtered = df_long[(df_long['Loja'] == loja) & (df_long['M√™s'] <= mes)]
        data.append(go.Scatter(x=df_filtered['M√™s'], y=df_filtered['Vendas'], mode='lines', name=loja))
    frames.append({'data': data, 'name': mes})

fig.frames = frames
```

## Definindo o layout e a anima√ß√£o




```python
fig.update_layout(
    title='An√°lise das Vendas Mensais por Loja em 2022',
    xaxis_title='M√™s',
    yaxis_title='Total de Vendas',
    updatemenus=[{
        'buttons': [
            {
                'args': [None, {'frame': {'duration': 500, 'redraw': False}, 'fromcurrent': True}],
                'label': 'Play',
                'method': 'animate'
            }
        ],
        'direction': 'left',
        'pad': {'r': 10, 't': 87},
        'showactive': False,
        'type': 'buttons',
        'x': 0.1,
        'xanchor': 'right',
        'y': 0,
        'yanchor': 'top'
    }],
    sliders=[{
        'steps': [
            {
                'args': [[f.name], {'frame': {'duration': 500, 'redraw': False}, 'mode': 'immediate', 'transition': {'duration': 300, 'easing': 'cubic-in-out'}}],
                'label': f.name,
                'method': 'animate'
            } for f in fig.frames
        ],
        'transition': {'duration': 300, 'easing': 'cubic-in-out'},
        'x': 0.1,
        'pad': {'b': 10, 't': 50},
        'len': 0.9
    }]
)
```

## Exibindo o gr√°fico

```python
fig.show()
```
